```python
import re
```

```python
def find_single_quotes(text):
    # Define a regular expression pattern to find single quotes and surrounding characters
    pattern = re.compile(r'\s+([^\s\']*\'[^\s\']*)\s+')

    # Find all matches in the text
    matches = re.findall(pattern, text)

    # Extract the content surrounded by single quotes
    results = [match.strip() for match in matches]

    return results
```
